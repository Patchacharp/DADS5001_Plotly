Import library

```python
import requests
import plotly.express as px
import pandas as pd
import time
import plotly.graph_objects as go
```

เลือกเหรียญหลักเป็นเหรียญ Bitcoin โดยใช้กราฟ 4 ชั่วโมง

```python
symbol = 'BTCUSDT&type=MINI&windowSize=4h'
api = "https://api.binance.com/api/v3/ticker?symbol="+symbol

BTC = []
for i in range(100):
    data = requests.get(api).json()
    BTC.append(data)
    time.sleep(1)

df_BTC = pd.DataFrame(BTC)
df_BTC.head(10)
```

เหรียญเปรียบเทียบคือเหรียญ Solana โดยใช้กราฟ 4 ชั่วโมง

```python
symbol = 'SOLUSDT&type=MINI&windowSize=4h'
api = "https://api.binance.com/api/v3/ticker?symbol="+symbol

SOL=[]
for i in range (100):
    data = requests.get(api).json()
    time.sleep(1)
    SOL.append(data)   

df_SOL = pd.DataFrame(SOL)
df_SOL.head(10)
```

สร้างกราฟ เปรียบเทียบระหว่างราคา และปริมาณการซื้อขายของทั้ง 2 เหรียญ

```python
Price_BTC = df_BTC["lastPrice"]
Price_SOL = df_SOL["lastPrice"]
Volume_BTC = df_BTC["quoteVolume"]
Volume_SOL = df_SOL["quoteVolume"]

fig = go.Figure()

# สร้างและปรับสไตล์ traces สำหรับราคา
fig.add_trace(go.Scatter(y=Price_BTC, name='BTC Price',
                         line=dict(color='firebrick', width=4)))
fig.add_trace(go.Scatter(y=Price_SOL, name='SOL Price',
                         line=dict(color='royalblue', width=4)))

# สร้างและปรับสไตล์ traces สำหรับปริมาณการซื้อขาย
fig.add_trace(go.Scatter(y=Volume_BTC, name='BTC Volume',
                         line=dict(color='firebrick', width=4, dash='dash'), yaxis="y2"))
fig.add_trace(go.Scatter(y=Volume_SOL, name='SOL Volume',
                         line=dict(color='royalblue', width=4, dash='dash'), yaxis="y2"))
# Edit the layout โดยกำหนดให้มีแกน y 2ค่า
fig.update_layout(
    title='Price and Volume Comparison of BTC and SOL',
    xaxis_title='Data Point (4hr)',
    yaxis_title='Price',
    yaxis=dict(
        title='Price'
    ),
    yaxis2=dict(
        title='Volume'
        ,
        overlaying='y',
        side='right'
    ),
    #ปรับสเกลและตำแหน่งของ legend
    legend=dict(
        x=1,
        y=1.35,
        traceorder='reversed',
        font=dict(
            size=10
        )
    )
)

fig.show()
```
